## Struktura i wgląd do danych

```python
df = pd.read_csv('dane_grib_era5.csv', index_col='time', parse_dates=True)
df.head(25)
df.tail(400)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.d2m.describe()
```

```python
df.sp = df.sp/100
df.msl = df.msl/100
df.d2m = df.d2m - 273.15
df.t2m = df.t2m - 273.15
df.tp = df.tp * 1000
df['wv'] = np.sqrt(df.u10**2 + df.v10**2)

df.describe()
```

```python
cols = ['sp', 'msl', 'tcc', 'wv', 't2m', 'd2m', 'ssrd', 'tp']
fig, axes = plt.subplots(2, 4, figsize=(10, 7))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.boxplot(data=df, y=col, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Rozkład zmiennej {col}')
    axes[i].set_xlabel('')

plt.tight_layout()
plt.show()
```

```python
cols = ['sp', 'msl', 'tcc', 'wv', 't2m', 'd2m', 'ssrd', 'tp']

fig, axes = plt.subplots(3,3, figsize=(9,7))
axes = axes.flatten()

for i, var in enumerate(cols):
  sns.histplot(data=df, x=var, ax=axes[i], bins=25, color='green')
  axes[i].set_title(f'Rozkład: {var}')

plt.tight_layout()
plt.show()
```

```python
df['valid_time'] = pd.to_datetime(df['valid_time'])
summer_months = [6, 7, 8]
winter_months = [12,1,2]
summer_6am = df[
    (df['valid_time'].dt.month.isin(summer_months)) &
    (df['valid_time'].dt.hour == 6)
]
print(summer_6am[['valid_time', 'ssrd']].head(10))

summer_0am = df[
    (df['valid_time'].dt.month.isin(summer_months)) &
    (df['valid_time'].dt.hour == 0)
]
print(summer_0am[['valid_time', 'ssrd']].head(10))

winter_6am = df[
    (df['valid_time'].dt.month.isin(winter_months)) &
    (df['valid_time'].dt.hour == 6)
]
print(winter_6am[['valid_time', 'ssrd']].head(10))
```

## Analiza trendów, sezonowości i przebiegu danych (w poszczególnych latach oraz w całym cyklu)

```python
cols = ['sp', 'msl', 'tcc', 'wv', 't2m', 'd2m', 'ssrd', 'tp']

year_means = df.groupby(df['valid_time'].dt.year)[cols].mean()
years = df['valid_time'].dt.year
fig, axes = plt.subplots(3,3,figsize=(10, 8))
axes = axes.flatten()

for i, var in enumerate(cols):
  sns.lineplot(data=year_means, x=year_means.index, y=var, ax=axes[i], color='green')
  axes[i].set_title(f'Roczne wartości średnie {cols[i]} dla lat 2010-2024', fontsize=8)
  axes[i].set(xlabel='Rok')
plt.tight_layout()
plt.show()
```

```python
cols = ['sp', 'msl', 'tcc', 'wv', 't2m', 'd2m', 'ssrd', 'tp']

cor_matrix = df[cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(cor_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title("Macierz współczynników korelacji liniowej Pearsona")
plt.show()
```

```python
sample_df = df.sample(1000)
sns.pairplot(sample_df[['t2m', 'd2m', 'sp', 'ssrd', 'tp']], kind='scatter', diag_kind='kde', plot_kws={'alpha': 0.5})
plt.suptitle("Zależności między zmiennymi w losowej próbce", y=1.02)
plt.show()
```

## Analiza ekstremów

```python
cols = ['sp', 'msl', 'tcc', 'wv', 't2m', 'd2m', 'ssrd', 'tp']

fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20,14))
axes = axes.flatten()

for i, col in enumerate(cols):
  ax = axes[i]
  kwantyl95 = df[col].quantile(0.95)
  extrema = df[df[col] > kwantyl95]

  yearly_counts = extrema.groupby(extrema['valid_time'].dt.year)[col].count()

  all_years = df['valid_time'].dt.year.unique()
  yearly_counts = yearly_counts.reindex(all_years, fill_value=0).sort_index()

  yearly_counts.plot(kind='bar', ax=ax, color='crimson', alpha=0.7)

  ax.set_title(f'Dni ekstremalne: {col}\n(Próg > {kwantyl95:.2f})', fontsize=8)
  ax.set_ylabel('Liczba dni w roku', fontsize=6)
  ax.set_xlabel('Rok', fontsize=6)
  ax.grid(axis='y', alpha=0.3)
```

```python
cols = ['sp', 'msl', 'tcc', 'wv', 't2m', 'd2m', 'ssrd', 'tp']


df_2015 = df[df['valid_time'].dt.year == 2015]
df_rest = df[df['valid_time'].dt.year != 2015]

corr_2015 = df_2015[cols].corr()
corr_rest = df_rest[cols].corr()


corr_diff = corr_2015 - corr_rest


fig, axes = plt.subplots(1, 3, figsize=(24, 7))

sns.heatmap(corr_rest, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1, ax=axes[0])
axes[0].set_title('Norma (Lata  bez 2015)')

sns.heatmap(corr_2015, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1, ax=axes[1])
axes[1].set_title('Anomalia (Tylko rok 2015)')

sns.heatmap(corr_diff, annot=True, fmt=".2f", cmap='PiYG', center=0, ax=axes[2])
axes[2].set_title('RÓŻNICA (Zmiana relacji w 2015)')

plt.tight_layout()
plt.show()
```

```python
year_means = df.groupby(df['valid_time'].dt.year)[cols].mean()
df_2015 = df[df['valid_time'].dt.year == 2015]
monthly_2015 = df_2015.groupby(df_2015['valid_time'].dt.month)[cols].mean()


fig = plt.figure(figsize=(20, 10), layout='constrained')
subfigs = fig.subfigures(1, 2, wspace=0.07)

subfigs[0].suptitle('Lata 2010-2024 (Trendy roczne)', fontsize=16)
subfigs[1].suptitle('Rok 2015 (Przebieg miesięczny)', fontsize=16)

axes_left = subfigs[0].subplots(4, 2).flatten()
axes_right = subfigs[1].subplots(4, 2).flatten()


for i, var in enumerate(cols):
    sns.lineplot(data=year_means, x=year_means.index, y=var, ax=axes_left[i], color='green')
    axes_left[i].set_title(f'{var}', fontsize=10)
    axes_left[i].set_xlabel('')
    axes_left[i].grid(True, alpha=0.3)

    sns.lineplot(data=monthly_2015, x=monthly_2015.index, y=var, ax=axes_right[i], color='red')
    axes_right[i].set_title(f'{var}', fontsize=10)
    axes_right[i].set_xlabel('Miesiąc')
    axes_right[i].set_xticks(range(1, 13))
    axes_right[i].grid(True, alpha=0.3)

plt.show()
```

```python
year_means = df.groupby(df['valid_time'].dt.year)[cols].mean()
df_2024 = df[df['valid_time'].dt.year == 2024]
monthly_2024 = df_2024.groupby(df_2024['valid_time'].dt.month)[cols].mean()


fig = plt.figure(figsize=(20, 10), layout='constrained')
subfigs = fig.subfigures(1, 2, wspace=0.07)

subfigs[0].suptitle('Lata 2010-2024 (Trendy roczne)', fontsize=16)
subfigs[1].suptitle('Rok 2024 (Przebieg miesięczny)', fontsize=16)

axes_left = subfigs[0].subplots(4, 2).flatten()
axes_right = subfigs[1].subplots(4, 2).flatten()


for i, var in enumerate(cols):
    sns.lineplot(data=year_means, x=year_means.index, y=var, ax=axes_left[i], color='green')
    axes_left[i].set_title(f'{var}', fontsize=10)
    axes_left[i].set_xlabel('')
    axes_left[i].grid(True, alpha=0.3)

    sns.lineplot(data=monthly_2024, x=monthly_2024.index, y=var, ax=axes_right[i], color='red')
    axes_right[i].set_title(f'{var}', fontsize=10)
    axes_right[i].set_xlabel('Miesiąc')
    axes_right[i].set_xticks(range(1, 13))
    axes_right[i].grid(True, alpha=0.3)

plt.show()
```

Ile anomalii temperatur?

```python
daily_max = df['t2m'].resample('D').max()
hot_days = (daily_max > 30).resample('Y').sum()

plt.figure(figsize=(12, 6))
bars = plt.bar(hot_days.index.year, hot_days.values, color='orange', alpha=0.8, edgecolor='red')

plt.title('Liczba dni upalnych >30 st. w roku')
plt.ylabel('Liczba dni')
plt.xlabel('Rok')
plt.grid(axis='y', linestyle='--', alpha=0.5)

for bar in bars:
    yval = bar.get_height()
    if yval > 0:
        plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, int(yval), ha='center', va='bottom')

plt.show()
```

Ile anomalii całkowitego opadu?

```python
annual_precip = df['tp'].resample('Y').sum()
norm_period = annual_precip['2010':'2024']
average_precip = norm_period.mean()

print(f"Norma roczna opadów wynosi: {average_precip:.2f} mm")

precip_anomaly = annual_precip - average_precip

plt.figure(figsize=(12, 6))

colors = ['brown' if x < 0 else 'green' for x in precip_anomaly]
plt.bar(precip_anomaly.index.year, precip_anomaly.values, color=colors, alpha=0.7)

plt.title(f'Roczne odchylenie opadów od normy (Norma = {average_precip:.0f} mm)')
plt.ylabel('Nadmiar / Niedobór opadów [mm]')
plt.axhline(0, color='black', linewidth=1)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```
